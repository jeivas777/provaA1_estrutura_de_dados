```python
import pandas as pd
# 1. Criação do DataFrame com dados de filmes
dados_filmes = {
    'nome_do_filme': ['A Origem', 'Batman: O Cavaleiro das Trevas', 'Pulp Fiction: Tempo de Violência', 'Matrix', 'Forrest Gump: O Contador de Histórias',
                      'O Senhor dos Anéis: A Sociedade do Anel', 'Gladiador', 'Titanic', 'Avatar', 'O Rei Leão'],
    'nome_original': ['Inception', 'The Dark Knight', 'Pulp Fiction', 'The Matrix', 'Forrest Gump',
                      'The Lord of the Rings: The Fellowship of the Ring', 'Gladiator', 'Titanic', 'Avatar', 'The Lion King'],
    'nome_do_diretor': ['Christopher Nolan', 'Christopher Nolan', 'Quentin Tarantino', 'The Wachowskis', 'Robert Zemeckis',
                        'Peter Jackson', 'Ridley Scott', 'James Cameron', 'James Cameron', 'Roger Allers, Rob Minkoff'],
    'bilheteria': [829895144, 1004558444, 213928762, 466364845, 678222284,
                   871530324, 457640427, 2187463944, 2847246203, 968511805],
    'orcamento': [160000000, 185000000, 8000000, 63000000, 55000000,
                  93000000, 103000000, 200000000, 237000000, 45000000],
    'ano_de_lancamento': [2010, 2008, 1994, 1999, 1994,
                          2001, 2000, 1997, 2009, 1994],
    'genero': ['Sci-Fi', 'Action', 'Crime', 'Sci-Fi', 'Drama',
               'Fantasy', 'Action', 'Drama', 'Sci-Fi', 'Animation']
}

# Criar o DataFrame
df_filmes = pd.DataFrame(dados_filmes)

# Exibir o DataFrame
print(df_filmes)
```

```python
# 2. Número de linhas e colunas
numero_linhas, numero_colunas = df_filmes.shape
print("numero de linhas: ", numero_linhas)
print("numero de colunas: ", numero_colunas)
```

```python
# 3. Analisando um item específico
print(df_filmes.iloc[1])
```

```python
# 4. Verificando se o DataFrame está vazio
if df_filmes.empty:
    print("O DataFrame está vazio.")
else:
    print("O DataFrame não está vazio.")
```

```python
# 5. Mostrando em tela os 5 primeiros itens do DataFrame
print(df_filmes.head())
```

```python
# 6. Deletando uma linha específica do DataFrame
df_filmes = df_filmes.drop(1)

# Exibir o DataFrame após as alterações
print(df_filmes)
```

```python
# 7. Adicionando um novo filme ao DataFrame

novo_filme = {
    'nome_do_filme': 'Vingadores: Ultimato',
    'nome_original': 'Avengers: Endgame',
    'nome_do_diretor': 'Anthony Russo, Joe Russo',
    'bilheteria': 2797501328,
    'orcamento': 356000000,
    'ano_de_lancamento': 2019,
    'genero': 'Action'
}

df_filmes = df_filmes.append(novo_filme, ignore_index=True)

print(df_filmes)
```

```python
# 8. Transpondo a coluna para a linha no DataFrame

df_filmes_transposto = df_filmes.transpose()
print(df_filmes_transposto)
```

```python
# 9. Mostrando na tela apenas as 2 primeiras colunas

df_filmes[['nome_do_filme', 'nome_original']]
```
